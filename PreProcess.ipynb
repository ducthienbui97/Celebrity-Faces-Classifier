```python
import numpy as np
import os
from keras.preprocessing.image import ImageDataGenerator, load_img,img_to_array
```

```python
train_dir = os.path.join('data','train')
for label in os.listdir(train_dir):
    label_dir = os.path.join(train_dir,label)
    all_img = os.listdir(label_dir)
    idx = len(all_img)
    datagen = ImageDataGenerator(
                    shear_range=0.2,
                    zoom_range=0.2,
                    horizontal_flip=True)
    img_set = []
    for image in all_img:
        img = img_to_array(load_img(os.path.join(label_dir,image)))
        img_set.append(img)
    for x in datagen.flow(np.asarray(img_set),
                          batch_size = 32,
                          save_to_dir = label_dir,
                          save_prefix = label,
                          save_format='jpeg'):
        idx += 32
        if idx > 1500:
            break
    print(label,len(os.listdir(label_dir)))
```

```python
validate_dir = os.path.join('data','validate')
for label in os.listdir(validate_dir):
    label_dir = os.path.join(validate_dir,label)
    all_img = os.listdir(label_dir)
    idx = len(all_img)
    datagen = ImageDataGenerator(
                    shear_range=0.2,
                    zoom_range=0.2,
                    horizontal_flip=True)
    img_set = []
    for image in all_img:
        img = img_to_array(load_img(os.path.join(label_dir,image)))
        img_set.append(img)
    for x in datagen.flow(np.asarray(img_set),
                          batch_size = 32,
                          save_to_dir = label_dir,
                          save_prefix = label,
                          save_format='jpeg'):
        idx += 32
        if idx > 300:
            break
    print(label,len(os.listdir(label_dir)))
```
